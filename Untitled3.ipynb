```python
import pandas as pd
import torch
from torch.utils.data import DataLoader, Dataset
from transformers import AutoTokenizer, AutoModelForSequenceClassification, Trainer, TrainingArguments
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from tqdm import tqdm

# Vérification de l'existence de GPU dans Google Colab
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Charger la dataset
# Assurez-vous que vous avez téléchargé le fichier Enron.csv dans le répertoire "BERT" sur Google Colab
file_path = '/content/BERT/Enron.csv'  # Remplacez ce chemin si nécessaire
df = pd.read_csv(file_path)

# Nettoyer et préparer les données (combinaison du sujet et du corps)
df['text'] = df['subject'] + " " + df['body']
df = df[['text', 'label']]  # Garder les colonnes nécessaires

# Fonction de nettoyage du texte
def clean_text(text):
    return str(text).lower()

df['text'] = df['text'].apply(clean_text)

# Diviser les données en jeu d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(df['text'], df['label'], test_size=0.2, random_state=42)

# Charger le tokenizer et le modèle pré-entraîné (distilBERT pour sa légèreté)
MODEL_NAME = 'distilbert-base-uncased'
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)

# Tokenisation des données
train_encodings = tokenizer(list(X_train), truncation=True, padding=True)
test_encodings = tokenizer(list(X_test), truncation=True, padding=True)

# Dataset PyTorch pour utiliser avec DataLoader
class EmailDataset(Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item['labels'] = torch.tensor(self.labels.iloc[idx], dtype=torch.long)
        return item

    def __len__(self):
        return len(self.labels)

train_dataset = EmailDataset(train_encodings, y_train)
test_dataset = EmailDataset(test_encodings, y_test)

# Charger le modèle BERT pour la classification
model = AutoModelForSequenceClassification.from_pretrained(MODEL_NAME, num_labels=2)
model.to(device)

# Définir les paramètres d'entraînement
training_args = TrainingArguments(
    output_dir='./results',          # Dossier pour les résultats
    num_train_epochs=3,              # Nombre d'époques
    per_device_train_batch_size=16,  # Taille du batch d'entraînement
    per_device_eval_batch_size=64,   # Taille du batch pour l'évaluation
    warmup_steps=500,                # Nombre de warmup steps
    weight_decay=0.01,               # Taux de régularisation
    logging_dir='./logs',            # Dossier pour les logs
    logging_steps=10,                # Fréquence de logging
    evaluation_strategy="epoch",     # Évaluation après chaque époque
    save_strategy="epoch",           # Sauvegarde après chaque époque
)

# Initialiser le Trainer
trainer = Trainer(
    model=model,                         # Le modèle pré-entraîné
    args=training_args,                  # Les arguments d'entraînement
    train_dataset=train_dataset,         # Dataset d'entraînement
    eval_dataset=test_dataset            # Dataset de test
)

# Lancer l'entraînement
trainer.train()

# Évaluation du modèle sur le jeu de test
results = trainer.evaluate()
print("Test Results:", results)

# Rapport de classification
y_pred = trainer.predict(test_dataset).predictions.argmax(axis=-1)
print(classification_report(y_test, y_pred))
```

```python
# Sauvegarder le modèle après l'entraînement
model.save_pretrained("./phishing_BERT_model")
tokenizer.save_pretrained("./phishing_BERT_model")
```

```python
# Exemple de nouveaux emails à classifier
new_emails = ["Congratulations! You've won a free iPhone!", "Reminder: Your meeting is scheduled for tomorrow at 3 PM."]

# Tokenisation des nouveaux emails
new_encodings = tokenizer(new_emails, truncation=True, padding=True, max_length=512, return_tensors="pt")
new_encodings = {key: val.to(device) for key, val in new_encodings.items()}
```

```python
# Effectuer les prédictions
model.eval()
with torch.no_grad():
    outputs = model(**new_encodings)
    logits = outputs.logits
    predictions = torch.argmax(logits, dim=1)

# Afficher les résultats (0 = non-phishing, 1 = phishing)
for i, email in enumerate(new_emails):
    print(f"Email: {email}")
    print(f"Prediction: {'Phishing' if predictions[i] == 1 else 'Non-Phishing'}")
```
